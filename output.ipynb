# Rock Paper Scissor Classifier

### Before running this program make sure that background is absolutely white

```python
import tensorflow as tf
import cv2
import time
import numpy as np
import matplotlib.pyplot as plt

print(tf.__version__)
```

```python
model = tf.keras.models.load_model("rps.h5")
labels = ["Paper", "Rock", "Scissor"]
```

```python

print("Capturing photo SMILE......")
time.sleep(5)
cam = cv2.VideoCapture(0)
cam.set(3, 320)
cam.set(4, 240)
check, frame = cam.read()

cam.release()
plt.imshow(frame)
  
pred_gen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=(1/255.0))
c = frame
new_c_r = c.reshape(1,c.shape[0],c.shape[1],c.shape[2])
pred_img = pred_gen.flow(new_c_r)
prediction = model.predict(pred_img)
print(prediction)
print(labels[np.argmax(prediction)])

cv2.destroyAllWindows()
```
